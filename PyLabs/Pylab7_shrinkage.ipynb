### Ridge Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
from sklearn.linear_model import LassoCV

from sklearn.metrics import mean_squared_error
```

```python
bd = pd.read_csv("https://raw.githubusercontent.com/neylsoncrepalde/ML_classes/master/Data/Hitters.csv")
bd.head()
```

```python
# Preprocessing
bd = pd.get_dummies(bd, drop_first=True)
bd = bd.dropna()
bd.shape
```

```python
# Objetivo: predizer o Sal√°rio (Salary)
y = bd.Salary
X = bd.drop(columns="Salary")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=2)
```

```python
rid = RidgeCV(alphas=np.linspace(0.1, 10), fit_intercept=True, normalize=True, 
              cv=None, # Use eficient LOOCV
             store_cv_values=True)
rid.fit(X_train, y_train)
```

```python
rid.alpha_
```

```python
rid.coef_
```

```python
rid.cv_values_
```

```python
yhat_rid = rid.predict(X_test)
np.sqrt(mean_squared_error(y_true=y_test, y_pred=yhat_rid))
```

### LASSO

```python
lasso = LassoCV(n_alphas=100, normalize=True, cv=10)
```

```python
lasso.fit(X_train, y_train)
```

```python
lasso.alphas_
```

```python
lasso.alpha_
```

```python
lasso.coef_
```

```python
lasso.mse_path_
```

```python
yhat_lasso = lasso.predict(X_test)
np.sqrt(mean_squared_error(y_test, yhat_lasso))
```
